```python
import os
from onekey_algo.regression.run_regression import main as reg_main
from collections import namedtuple
from onekey_algo import get_param_in_cwd

# 设置参数
task_type = 'L'
root = get_param_in_cwd('radio_dir')
for cv in range(0, 1):
    for model_name in ['resnet50', 'resnet18', 'resnet101', 'densenet121', 'vgg19', 'inception_v3']:
        train_f = os.path.join('split_info', f'train-RND-{cv}.txt')
        val_f = os.path.join('split_info', f'val-RND-{cv}.txt')
        labels_f = os.path.join('split_info', 'labels.txt')
        data_pattern = os.path.join(root, task_type)
        params = dict(train=train_f,
                      valid=val_f,
                      labels_file=labels_f,
                      in_channels=3,
                      data_pattern=data_pattern,
                      j=6,
                      max2use=None,
                      val_max2use=None,
                      batch_balance=False,
                      normalize_method='imagenet',
                      model_name=model_name,
                      gpus=[0],
                      batch_size=16,
                      epochs=64 if model_name not in ['TwinsSVT', 'CrossFormer', 'SimpleViT'] else 96,
                      init_lr=0.001 if model_name not in ['TwinsSVT', 'CrossFormer', 'SimpleViT'] else 0.001,
                      optimizer='sgd',
                      retrain='',
                      model_root= os.path.join(root, 'Lmodels', f"CV-{cv}"),
                      iters_start=0,
                      iters_verbose=4,
                      save_per_epoch=True,
                      add_date=False,
                      pretrained=True)
        # 训练模型
        Args = namedtuple("Args", params)
        reg_main(Args(**params))
```
